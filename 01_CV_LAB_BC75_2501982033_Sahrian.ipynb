```python
path = '/content/Test.jpg'
im = Image.open(path)
print(im.width, im.height, im.mode, im.format, type(im))
```

```python
plt.figure(figsize = (5,5))
plt.imshow(im)
plt.show()
```

```python
im_g = im.convert('I')
```

```python
plt.figure(figsize = (5,5))
plt.imshow(im_g)
plt.show()
```

```python
def compare_image(image1, image2):
 plt.figure(figsize = (9,9))
 plt.subplot(1,2,1)
 plt.imshow(image1)
 plt.title('Original')

 plt.subplot(1,2,2)
 plt.imshow(image2)
 plt.title('Modified')

 plt.tight_layout()
```

```python
compare_image(im, im_g)
```

```python
im_c = im.crop((400, 200, 600, 400))

plt.figure(figsize = (5,5))
plt.imshow(im_c)

plt.show()
```

```python
im_large = im.resize((im.width*2, im.height*2,), Image.BILINEAR)

print(im_large.width, im_large.height)

plt.figure(figsize = (5,5))
plt.imshow(im_large)
plt.show()
```

```python
im_small = im.resize((im.width//10, im.height//10,), Image.BILINEAR)

print(im_small.width, im_small.height)

plt.figure(figsize = (5,5))
plt.imshow(im_small)
plt.show()
```

```python
im_45 = im.rotate(45)

compare_image(im, im_45)
```
